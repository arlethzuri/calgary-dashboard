```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
import json
import pandas as pd
from shapely.geometry import shape
```

```python
# Open the file with geopandas and print its geometry type
data_path = "/sci-it/hosts/olympus/calgary/data/open_calgary/20260327/7t9h-2z9s/7t9h-2z9s_data.json"
metadata_path = "/sci-it/hosts/olympus/calgary/data/open_calgary/20260327/7t9h-2z9s/7t9h-2z9s_metadata.json"
# load json file, an iterable object of GeoJSON feature(s)
with open(data_path, 'r') as f:
    data_obj = json.load(f)

# unwrap data_obj
data_obj = data_obj[0]
with open(metadata_path, 'r') as f:
    metadata_obj = json.load(f)

# projection = metadata_obj["metadata"]["custom_fields"]["Geospatial Information"]["Map Projection"]

# data_obj[0].keys()
```

```python
def get_geometry_type(data_json_obj):
    """
    Get the geometry type from the Open Calgary dataset JSON.
    """
    geometry_types = ['point', 'linestring', 'polygon', 'multipoint', 'multilinestring', 'multipolygon']
    data_json_obj_keys = [k.lower() for k in data_json_obj.keys()]
    for geometry_type in geometry_types:
        if geometry_type in data_json_obj_keys:
            return geometry_type
    return None
```

```python
# get GeoJSON geometry type from data file
geometry_type = get_geometry_type(data_obj)
geometry_type
```

```python
# basic plot of enmax as parquet data

dir = "/sci-it/hosts/olympus/calgary/processed_data/enmax/20260327/features/"

parquet_files = [f for f in os.listdir(dir) if f.endswith('.parquet')]

for pq_file in parquet_files:
    gdf = gpd.read_parquet(os.path.join(dir, pq_file))
    fig, ax = plt.subplots(figsize=(10, 10))
    gdf.plot(ax=ax)
    ax.set_title(pq_file)
    plt.show()
```
